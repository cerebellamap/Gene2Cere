```python
import time
import tracemalloc
from memory_profiler import memory_usage
    
def my_function():
    # 示例代码
    result = [x**2 for x in range(10000)]
    return result

# 测量内存使用情况
mem_usage = memory_usage(my_function)
print(f"Memory usage: {max(mem_usage) - min(mem_usage)} MiB")

# 测量执行时间
start_time = time.time()
my_function()
end_time = time.time()
print(f"Execution time: {end_time - start_time} seconds")

# 使用 tracemalloc
tracemalloc.start()
my_function()
snapshot = tracemalloc.take_snapshot()
top_stats = snapshot.statistics('lineno')

print("[ Top 10 memory usage ]")
for stat in top_stats[:10]:
    print(stat)
```

```python
import os

def display_path_structure(path, indent=0, max_depth=2):
    if indent < max_depth:
        items = sorted(os.listdir(path))
        for item in items:
            item_path = os.path.join(path, item)
            print("    " * indent + "|-- " + item)
            if os.path.isdir(item_path):
                display_path_structure(item_path, indent + 1, max_depth)

# 示例用法
path_to_display = "/n02dat01/users/ypwang/Gradient/STARProtocol/Try"  # 将此处替换为你想要展示的路径
print(path_to_display)
display_path_structure(path_to_display)
```

```python
import psutil
from memory_profiler import profile
@profile
def run_brainsmash(input_file, n_permutations):
    print_system_memory("Before Step02_Brainsmash")
    G2C.Step02_Brainsmash(input_file, n_permutations)
    print_system_memory("After Step02_Brainsmash")

    print_system_memory("Before Step02_Brainsmash2IDP")
    G2C.Step02_Brainsmash2IDP(n_permutations)
    print_system_memory("After Step02_Brainsmash2IDP")

    print_system_memory("Before Step02_Brainsmash2PLSR")
    preds_name, score_name = G2C.Step02_Brainsmash2PLSR(n_components=8, median_index=20, n_permutations=n_permutations)
    print_system_memory("After Step02_Brainsmash2PLSR")

    print_system_memory("Before Step02_Model_eval_visualization")
    G2C.Step02_Model_eval_visualization(n_components=8, cv_repeat=101, median_index=20, median_score=0.4035226203938596)
    print_system_memory("After Step02_Model_eval_visualization")

def print_system_memory(step):
    # 获取虚拟内存信息
    memory_info = psutil.virtual_memory()
    print(f"{step} - Total RAM: {memory_info.total / (1024**3):.2f} GB")
    print(f"{step} - Available RAM: {memory_info.available / (1024**3):.2f} GB")
    print(f"{step} - Used RAM: {memory_info.used / (1024**3):.2f} GB")
    print(f"{step} - Free RAM: {memory_info.free / (1024**3):.2f} GB")

if __name__ == "__main__":
    run_brainsmash(input_file='./Data/Input_example/Input_example.nii', n_permutations=1000)
```
